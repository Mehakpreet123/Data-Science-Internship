```python
import numpy as np
import matplotlib.pyplot as plt 

 
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30, 
        'Python':35}
courses = list(data.keys())
values = list(data.values())
 
fig = plt.figure(figsize = (7, 3))

# creating the bar plot
plt.bar(courses, values, color ='maroon', 
       width=0.3 )

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")

plt.show()
```


plt.subplots in Matplotlib is a utility function that creates a figure and a set of subplots in a single call. 
plt.subplots(nrows=2, ncols=2) creates a 2x2 grid of subplots.
The axes array allows you to access and manipulate each subplot individually.
plt.tight_layout() is used to automatically adjust the subplot parameters to give some padding.

```python


# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot data on each subplot
axes[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 0].set_title('First subplot')

axes[0, 1].plot([1, 2, 3, 4], [16, 9, 4, 1])
axes[0, 1].set_title('Second subplot')

axes[1, 0].plot([1, 2, 3, 4], [2, 3, 2, 3])
axes[1, 0].set_title('Third subplot')

axes[1, 1].bar([1, 2, 3, 4], [1, 2, 1, 2])
axes[1, 1].set_title('Fourth subplot')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

fig, ax:

fig: This is the Figure object which is the overall container for all the plot elements.
ax: This is the Axes object or the single subplot, where most of the plotting methods are called.

plt.subplots(): This function is used to create a figure and a grid of subplots. Since no nrows or ncols parameters are specified, it defaults to creating a single subplot.
figsize=(16, 9): This argument sets the size of the figure in inches. It means the figure will be 16 inches wide and 9 inches tall. This aspect ratio is commonly used for wide-screen displays and presentations.

```python

# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data,bins=30,  color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt 
# data to display on plots 
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

# This will plot a simple scatter chart
plt.scatter(x, y)

# Adding legend to the plot
plt.legend("A")

# Title to the plot
plt.title("Scatter chart")
plt.show()
```

```python

# List of Days
days = [1, 2, 3, 4, 5]

# No of Study Hours
Studying = [7, 8, 6, 11, 7]

# No of Playing Hours
playing = [8, 5, 7, 8, 13]

# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
            colors =['r', 'c'])

# Days
plt.xlabel('Days')

# No of hours
plt.ylabel('No of Hours')

# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')

# Displaying Graph
plt.show()
```

```python

# data to display on plots 
x = [1, 2, 3, 4] 

# this will explode the 1st wedge
# i.e. will separate the 1st wedge
# from the chart
e  =(0.1, 0, 0, 0)

# This will plot a simple pie chart
plt.pie(x, explode = e)

# Title to the plot
plt.title("Pie chart")
plt.show()
```
