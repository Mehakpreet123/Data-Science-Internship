Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

```python
import matplotlib.pyplot as plt
import numpy as np
```

The plot() function is used to draw points (markers) in a diagram.
By default, the plot() function draws a line from point to point.
The function takes parameters for specifying points in the diagram.
Parameter 1 is an array containing the points on the x-axis.
Parameter 2 is an array containing the points on the y-axis.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

# to plot without showing line

plt.plot(xpoints, ypoints, 'o')
plt.show()

# If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc.,
# depending on the length of the y-points.
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

MATPLOTLIB SUBPLOT

fig, axs = plt.subplots(2, 2, figsize=(10, 8)): Creates a figure with a 2x2 grid of subplots. figsize=(10, 8) specifies the size of the entire figure.
axs[0, 0].plot(x, y1, 'b-', label='sin(x)'): Plots y1 versus x in the first subplot (axs[0, 0]) with a blue line.
axs[0, 0].set_title('sin(x)'): Sets the title for the first subplot.
axs[0, 0].grid(True): Enables the grid for the first subplot.
axs[0, 0].legend(): Adds a legend to the first subplot.
fig.suptitle('Subplot Example', fontsize=16): Adds an overall title to the entire figure.
plt.tight_layout(rect=[0, 0.03, 1, 0.95]): Adjusts the layout to make room for the overall title.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create the figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on each subplot
axs[0, 0].plot(x, y1, 'b-', label='sin(x)')
axs[0, 0].set_title('sin(x)')
axs[0, 0].grid(True)
axs[0, 0].legend()

axs[0, 1].plot(x, y2, 'r-', label='cos(x)')
axs[0, 1].set_title('cos(x)')
axs[0, 1].grid(True)
axs[0, 1].legend()

axs[1, 0].plot(x, y3, 'g-', label='tan(x)')
axs[1, 0].set_title('tan(x)')
axs[1, 0].set_ylim(-10, 10) 
axs[1, 0].grid(True)
axs[1, 0].legend()

axs[1, 1].plot(x, y4, 'm-', label='exp(x)')
axs[1, 1].set_title('exp(x)')
axs[1, 1].set_ylim(0, 22000) 
axs[1, 1].grid(True)
axs[1, 1].legend()

# Add overall titles and adjust layout
fig.suptitle('Subplot Example', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Display the plot
plt.show()
```

SCATTER PLOT
np.random.seed(0): Sets the seed for NumPy's random number generator to ensure reproducibility of the random numbers.
np.random.rand(50): Generates 50 random numbers between 0 and 1 for both x and y.
colors = np.random.rand(50): Generates 50 random numbers for the color of each point.
sizes = 1000 * np.random.rand(50): Generates 50 random numbers for the size of each point, scaled up to make them more visible.
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis'): Creates a scatter plot with x and y coordinates. The c parameter specifies the color of each point, s specifies the size, alpha sets the transparency, and cmap specifies the colormap.
plt.xlabel('X-axis'): Adds a label to the x-axis.
plt.ylabel('Y-axis'): Adds a label to the y-axis.
plt.title('Scatter Plot Example'): Adds a title to the plot.
plt.colorbar(): Adds a color bar to the plot, showing the color scale.

```python

np.random.seed(0)  # For reproducibility
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)


plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Add a color bar
plt.colorbar()  

plt.show()
```

MATPLOT BARS :
edgecolor='black' and linewidth=1.2: Adds a black edge with a specific line width around each bar.
plt.grid(axis='y', linestyle='--', alpha=0.7): Adds horizontal grid lines with a dashed style and sets their transparency.
plt.text(i, value + 0.5, str(value), ha='center'): Adds the value of each bar as a text annotation slightly above the top of each bar, centered horizontally.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 17, 35, 29, 12]

plt.bar(categories, values, color='skyblue', edgecolor='black', linewidth=1.2)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Chart Example')


plt.grid(axis='y', linestyle='--', alpha=0.7)


for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center')

plt.show()
```

MATPLOTLIB HISTOGRAM:
np.random.seed(0): Sets the seed for NumPy's random number generator to ensure reproducibility of the random numbers.
np.random.randn(1000): Generates 1000 random numbers from a standard normal distribution (mean = 0, standard deviation = 1).
plt.hist(data, bins=30, color='skyblue', edgecolor='black'): Creates a histogram with the specified data. bins=30 specifies the number of bins (bars), color sets the color of the bars, and edgecolor sets the color of the edges of the bars.
plt.xlabel('Value'): Adds a label to the x-axis.
plt.ylabel('Frequency'): Adds a label to the y-axis.
plt.title('Histogram Example'): Adds a title to the plot.
plt.show(): Displays the plot.

```python
np.random.seed(0)  # For reproducibility
data = np.random.randn(1000)  # Generate 1000 random numbers from a standard normal distribution

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.show()
```

PIE CHART :
labels: List of category labels.
sizes: List of values representing the size of each category.
colors: List of colors for each slice of the pie chart.
explode: Tuple to specify if any slice should be "exploded" (i.e., pulled out slightly from the center). Here, the first slice ('A') is exploded.
plt.pie(...): Creates the pie chart with the specified data and customization options.
sizes: The values to plot.
explode: Which slices to explode.
labels: The labels for each slice.
colors: The colors for each slice.
autopct: A string to format the labels inside the pie slices as percentages.
shadow: Adds a shadow to the pie chart.
startangle: Rotates the start of the pie chart by the specified angle.
plt.title('Pie Chart Example'): Adds a title to the plot.
plt.show(): Displays the plot.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice (i.e. 'A')

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)


plt.title('Pie Chart Example')

plt.show()
```
