SERIES CREATION
 The object supports both integer and label-based indexing and provides a host of methods for performing operations involving the index. 

```python
#  Creating a series from array
data = np.array(['hi','hello','welcome','thankyou'])
 
ser = pd.Series(data, index=[10, 11, 12, 13]) # with index
print(ser)
print("-------------------------------------------------------")

# creating series from list
ser=pd.Series(["mehak","preet","Kaur"],index=['Name','Middle name','Last name'])
print(ser)
print("-------------------------------------------------------")
 
# creating series with dictionary
dict = {'Maths': 10,
        'English': 20,
        'History': 30}
 
# create series from dictionary
ser = pd.Series(dict) # keys will be used as index by default
 
print(ser)
print("-------------------------------------------------------")
# creating series with scalar value
ser = pd.Series(10, index=[0, 1, 2, 3, 4, 5])
 
print(ser)
print("-------------------------------------------------------")
# from tuple
ser = pd.Series((1,2,3))
print(ser)
```

```python
# using function
ser1 = pd.Series(np.linspace(3, 33, 3))
print(ser1)
print("-------------------------------------------------------")
# series with numpy linspace()
ser2 = pd.Series(np.linspace(1, 100, 10))
print(ser2)
print("-------------------------------------------------------")
# using arange
ser=pd.Series(np.arange(1,5))
print(ser)
print("-------------------------------------------------------")
# usinh range
import pandas as pd
ser=pd.Series(range(5))
print(ser)
print("-------------------------------------------------------")
# using expression
ser=np.arange(10,15)
serobj=pd.Series(data=ser*5,index=ser)
print(serobj)
print("-------------------------------------------------------")
# using loop
ser=pd.Series(range(1,20,3), index=[x for x in 'abcdefg'])
print(ser)
```

DATAFRAME CREATION
pandas.DataFrame(data, index, columns)

```python
# creation from list / ndarray
# creation from dictionary where keys are columns names and values are lists/series, each list is a full column
df=pd.DataFrame({"Name":["Mehak","Komal","Muskan"],"State":["Punjab","Punjab","Punjab"]})
print(df)
print("-------------------------------------------------------")

# using list of lists
data = [['tom', 10], ['nick', 15], ['juli', 14]] # each list is a row
# Create the pandas DataFrame
df = pd.DataFrame(data, columns=['Name', 'Age'])
print(df)

print("-------------------------------------------------------")
# from list of dictionaries

data = [{'a': 1, 'b': 2, 'c': 3},
        {'a': 10, 'b': 20, 'c': 30}]

df = pd.DataFrame(data) # each dictionary is a complete row with column name
print(df)
# passing row index also
data = [{'b': 2, 'c': 3}, {'a': 10, 'b': 20, 'c': 30}]

# Creates pandas DataFrame by passing
# Lists of dictionaries and row index.
df = pd.DataFrame(data, index=['first', 'second'])

# Print the data
print(df)
print("-------------------------------------------------------")
# Pandas Dataframe from Dicts of series.
d = {'one': pd.Series([10, 20, 30, 40],
                      index=['a', 'b', 'c', 'd']),
     'two': pd.Series([10, 20, 30, 40],
                      index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
print(df)
```

```python
# using zip

Name = ['tom', 'krish', 'nick', 'juli']

Age = [25, 30, 26, 22]

# get the list of tuples from two lists.
# and merge them by using zip().
list_of_tuples = list(zip(Name, Age))
df = pd.DataFrame(list_of_tuples, columns=['Name', 'Age'])
print(df)
```

SELECTION AND SLICING


```python
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
df=pd.DataFrame(data)
display(df)
name=df["Name"]
display(name)
display(df[['Name', 'Qualification']]) # list will be passed when we want to select multiple columns
```

```python
# using df.columns
simple_method=df[df.columns[1:4]] # will select the columns 1,2 and 3 
display(simple_method)

print(df.loc[2]) # to extract row by using the index (single row)
print(df.loc[:,"Age"]) # to extract one column
display(df.loc[[0,1,2]]) # multiple rows by passing list of row indices

#using loc[]
# select three rows and two columns
display(df.loc[1:3, ['Name', 'Qualification']]) # loc function goes from start to end

# select two rows and 
# column "name" to "Address"
# Means total three columns
display(df.loc[0:1, 'Name':'Address'])

# df.loc[rows, columns]
# all rows, all columns
display(df.loc[:, :])
# row 1, all columns
df.loc[0, :]
```

```python
# using iloc[] / df.ix[] is also there to use but depreciated

# Remember that Python does not
# slice inclusive of the ending index.
# select all rows 
# select first two column
display(df.iloc[:, 0:2] )

# iloc[row slicing, column slicing]
df.iloc [0:2, 1:3]
display(df.iloc[2]) # particular row 
display(df.iloc[:,2])  # particular column , it cannot take a label name can take up the int value only
display(df.iloc[[0,1,2]]) # multiple rows, pass list of indices
```

```python
# boolean Indexing
age_filter=df["Age"]>=24 # will return true false for every row
display(df[age_filter])

ind_filter=df.index>1
print(ind_filter)
df[ind_filter]
```
